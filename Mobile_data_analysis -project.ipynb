# 12. Which mobile has highest battery storage dual_sim or single_sim

```python
# Creating battery_range column

import pandas as pd
s=[]
for r in df["battery_power"]:
    if 1000>=r>=500:
        s.append("poor")
    elif 1500>=r>1000:
        s.append("average")
    else:
        s.append("good")
df["battery_range"]=pd.DataFrame({"battery_range":s})
df
```

```python
import numpy as np

r=[]
n=0
l=0
m=0
for k in df[df["dual_sim"]==1]["battery_range"]:
    if k=="poor":
        n+=1
    elif k=="average":
        l+=1
    else:
        m+=1
r.append(n)
r.append(l)
r.append(m)
          
v=[]
w=0
x=0
y=0

for k in df[df["dual_sim"]==0]["battery_range"]:
    if k=="poor":
        w+=1
    elif k=="average":
        x+=1
    else:
        y+=1

v.append(w)
v.append(x)
v.append(y)

import matplotlib.pyplot as plt
import pandas as pd 
divisions=[1,2,3]
dual_sim_batt=r
single_sim_batt=v
index=np.arange(3)
width=0.20
plt.bar(index,dual_sim_batt,width,color='b',label="Dual SIM Battery",tick_label=["poor","avg","good"])
plt.bar(index+width,single_sim_batt,width,color='r',label="Single SIM Battery")

plt.title("Dual and Single SIM battery usage")
plt.xlabel("Bttery range")
plt.ylabel("Battery level")
plt.legend(loc="best")
plt.show()
```

# 14. Screen size for dual_sim mobile phones with respect to the price_range 

```python
import numpy as np
import matplotlib.pyplot as plt

count=range(0,4)

for n in count:
    dflp=np.where((df["dual_sim"]==1) & (df["price_range"]==n))
    dflp1=df.loc[dflp]
    box_data=dflp1[["sc_h","sc_w"]]
    labels=["screen_height","screen_width"]
    color=['r','c']
    box=plt.boxplot(box_data,sym='p',labels=labels,patch_artist=True)
    for patch,color in zip(box['boxes'],color):
        patch.set_facecolor(color)
    plt.title(f"Graph of price_range {n}")
    plt.show()
```

# 15. Relate between screen hight and width

```python
import numpy as np
import matplotlib.pyplot as plt

x=df['sc_h'].sort_values().to_list()
y=df['sc_w'].sort_values().to_list()

firstx=x[:500]
firsty=y[:500]
secondx=x[500:1300]
secondy=y[500:1300]
thirdx=x[1300:]
thirdy=y[1300:]

Xvalues=firstx+secondx+thirdx
Yvalues=firsty+secondy+thirdy

cor=np.polyfit(Xvalues,Yvalues,1)
p=np.poly1d(cor)

# Set Values of color

col=['b']*len(firstx)+['g']*len(secondx)+['r']*len(thirdx)
# Plot values and line
plt.scatter(Xvalues,Yvalues,color=col)
plt.plot(Xvalues,p(Xvalues),'c--')
plt.title("Regression Line")
plt.xlabel("Screen_height")
plt.ylabel("Screen_width")
plt.grid()
plt.show()
```
